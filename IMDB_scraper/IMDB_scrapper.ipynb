```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import requests
import pandas as pd
```

```python
url = 'https://www.imdb.com/chart/top?ref_=nv_mv_250'
data = requests.get(url)
x = data.content
soup = BeautifulSoup(x, 'html.parser')
movie_table = soup.find('tbody', {'class': 'lister-list'})
movie_table
```

```python
lst_mov = movie_table.find_all('tr')
lst_mov
```

```python
movies =[]
ratings = []
for i in range(len(lst_mov)):
    ratings.append(lst_mov[i].find('td', {'class':"ratingColumn imdbRating"}).text[1:4])
    movies.append(lst_mov[i].find_all('a')[1].text)
```

```python
d = {'Movie': movies, 'Rating':ratings}
```

```python
data = pd.DataFrame(d)
data['Rank'] = list(data.index +1)
```

```python
data.to_excel("Movie_Ratings.xlsx")
```
